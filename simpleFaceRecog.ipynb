```python
#Here we are importing the OpenCV library and assigning it to the variable cv2. We create another variable called ‘facePath’ and assign it to the string
#'enter_face.jpg’. Our intention with this is to reference later when we want to specify the path to an image file which will be processed using OpenCV.
import cv2
facePath = "enter_face.jpg"
```

```python
#Here is where we read the image file from the specified imagePath and store it in the ‘img’ variable. ‘cv2.imread()’ is a function that’ll take the file
#path as an argument and return a NumPy array showing the image.
img = cv2.imread('Downloads/tim2.jpg')
```

```python
#Here we create an attribute called ‘shape’ for the variable ‘img’. Shape will return a tuple that has the dimensions of the chosen image. We get the 
#height, then width, and then the number of color channels in the image from the tuple.
img.shape
```

```python
#Here we are converting the original image to a grayscale image using the img variable. ‘cv2.cvtColor’ will take two arguments. Argument 1 is the image
#that will be converted and argument 2 is the code for color conversion. ‘cv2.COLOR_BGR2GRAY’ specifies that the image will be changed from the BGR to
#grayscale. The variable ‘gray_image’ stores the grayscale image.
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#This code returns the tuple for the dimensions of the ‘gray_image’. We only get the height and width for this image.
gray_image.shape
```

```python
#CascadeClassifier is a class in OpenCV that is used for object detection. The haarcascade_frontalface_default.xml file is a trained classifier that 
#detects frontal faces in image and videos. Here the CascadeClassifier is intitalized as an object ‘face_classifier’ to do this we use 
#haarcascade_frontalface_default.xml.file.
face_classifier = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
#We added the ‘detectMultiScale’ fuction to take the paramateres ‘gray_image’, ‘scaleFactor’ (to consider that faces may appear at different scales), 
#‘minNeighbors’ (specifies how many neighbors each detection rectangle should retain it), ‘minSize’ (minimum size of a face detected, ifnoring any face
#smaller that the size). All these values will be stored o=in the ‘face’ varaiable’. This function returns a list of rectangles that represent the 
#detected faces in the image.
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40,40)
)
```

```python
#Here we create the detection box. We created a for loop that will iterate over each face that is detected. A tuple of 4 values: x and y are the 
#coordinates of the top-left corner of the face rectangle, w is the width of the rectangle, and h is the height of the rectangle. These represent the 
#detected faces. The function cv2.rectangle() is called to place a rectangle on the original image (img) using the coordinates and dimensions of the 
#detected face.
for(x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 4)
```

```python
#We need to convert the image from BGR back to RGB so this is what we do by using the ‘cv2.cvtColor()’ function. The function takes two arguments, the 
#image to be converted and the color space conversion. The function returns a newly converted image (img_rgb).A lot of image processing libraries and 
#functions expect images to be in the RGB color space, so converting the image beforehand should prevent errors and ensure that the image is processed 
#correctly.
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#A new image is created through the module we imported ‘matplotlib.pyplot with a size of 20x10 (‘plt.figure()’). Using the ‘plt.imshow()’ function an 
#image ‘img_rgb’ is displayed. ‘plt.figure()’ creates a new figure with the specified size. ‘plt.imshow()’ displays an image on the figure. 
#‘plt.axis('off')’  turns off the axis, which is the set of lines that indicate the scale of the plot.
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
